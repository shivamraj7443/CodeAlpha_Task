## ðŸ“Œ Basic Questions for Exploratory Data Analysis

1. What is the average price of books?
2. What is the minimum and maximum price of books?
3. How many books are currently in stock?


**Average book price**

```python
df["Price"].mean()
```

**Minimum & Maximum price**

```python
df["Price"].min()
df["Price"].max()
```

**Availability count**

```python
df["Availability"].value_counts()
```

**Visual Analysis Price distribution (Histogram)**

```python
plt.hist(df["Price"])
plt.xlabel("Price")
plt.ylabel("Number of Books")
plt.title("Distribution of Book Prices")
plt.show()
```

**Price comparison (Bar â€“ first 10 books)**

```python
df.head(10).plot(x="Title", y="Price", kind="bar")
plt.xticks(rotation=90)
plt.title("Price of First 10 Books")
plt.show()
```

**Key Insights**

Most book prices fall within a specific range.

All books in the dataset are currently in stock.

There is variation in pricing across books.
