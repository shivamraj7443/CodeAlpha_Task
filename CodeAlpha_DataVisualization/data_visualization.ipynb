```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("books_data.csv")
df.head()
```

**Visualization 1 – Price Distribution (Histogram)**

```python
plt.hist(df["Price"])
plt.xlabel("Book Price")
plt.ylabel("Number of Books")
plt.title("Distribution of Book Prices")
plt.show()
```

**Visualization 2 – Price Comparison (Bar Chart)**

```python
df.head(10).plot(x="Title", y="Price", kind="bar")
plt.xlabel("Book Title")
plt.ylabel("Price")
plt.title("Price Comparison of First 10 Books")
plt.xticks(rotation=90)
plt.show()
```

**Insight:**

Prices vary significantly across books.

Some books are priced higher than others even within a small sample.

**Visualization 3 – Availability Count**

```python
df["Availability"].value_counts().plot(kind="bar")
plt.xlabel("Availability Status")
plt.ylabel("Count")
plt.title("Availability of Books")
plt.show()
```

**Insight:**

Majority of books are currently in stock.
